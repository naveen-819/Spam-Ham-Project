```python
import sys
import os
import pandas as pd
from sklearn.svm import LinearSVC
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
os.chdir("/Users/naveenrowthula/Downloads/")
```

```python
######READING DATA#######
train_data = pd.read_csv("spam_train.csv",encoding='latin-1',header=0)
test_data = pd.read_csv("spam_test.csv",encoding='latin-1',header=None)
train_v1=train_data['v1']
train_v2=train_data['v2']
test_v1=test_data[0]
test_v2=test_data[1]
```

```python
model = TfidfVectorizer(stop_words='english')
train_vector = model.fit_transform(train_v2)
```

```python
clf = LinearSVC()
clf.fit(train_vector,train_v1)
```

```python
test_vector = model.transform(test_v2)
pred = clf.predict(test_vector)
count = 0
for i in pred:
    if i=='spam':
        print(0)
    else:
        print(1)

for i,j in zip(test_v1,pred):
    if i==j:
        count =count+1
```

```python
#####ACCURACY#####

accuracy = (float(count)/len(test_v1))*100
print ('ACCURACY OBTAINED:',accuracy)
print("\n        THANK YOU")
```
